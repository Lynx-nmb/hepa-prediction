```python
x = n_data.drop(columns=['target'])
y = n_data['target']
```

```python
x.to_csv('x.csv', index=False)
```

```python
from category_encoders import BinaryEncoder
categorycal_features = ['gender','steroid', 'fatigue', 'liverBig', 'liverFirm']
encoder = BinaryEncoder()
x_encoded = encoder.fit_transform(x[categorycal_features])
x = x.drop(categorycal_features, axis=1)
x = x.reset_index(drop=True)
x_encoded = x_encoded.reset_index(drop=True)
x = pd.concat([x, x_encoded], axis=1)
```

```python
y.value_counts()
```

```python
# from sklearn.model_selection import train_test_split
# x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.25, random_state=43)
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()
x, y = smote.fit_resample(x, y)
```

```python
y.value_counts()
```

```python
from sklearn.metrics import accuracy_score, precision_score
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=.25, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier

dcs = DecisionTreeClassifier()
dcs.fit(x_train, y_train)
```

```python
# from sklearn.tree import DecisionTreeClassifier, export_graphviz
from joblib import dump, load
# import graphviz
dcs = DecisionTreeClassifier()
dcs.fit(x_train, y_train)

# export_graphviz(dcs, out_file='tree.dot', filled=True, feature_names=x_train.columns)

# with open('tree.dot') as f:
#     dot_graph = f.read()
# graphviz.Source(dot_graph)

dump(dcs, "decision_tree_model.joblib")
```

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score
)
prediction = dcs.predict(x_test)

accuracy = accuracy_score(prediction, y_test)
precision = precision_score(prediction, y_test)

print(f"Accuracy: {accuracy*100}")
print(f"Precision: {precision*100}")
```
